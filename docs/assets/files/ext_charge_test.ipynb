```python
import psi4
import numpy as np
```

```python
psi4.geometry( 
"""
H 0 0 1
H 0 0 0
symmetry c1
""")


external_potentials = [
    [-0.834, np.array([0, 0, 2]) ],
    [ 0.417, np.array([0, 0, 3]) ],
    [ 0.417, np.array([0, 0,.5]) ]]

psi4.energy('scf/sto-3g', external_potentials=external_potentials)
```
